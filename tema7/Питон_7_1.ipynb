```python
import matplotlib.pyplot as plt


x = list(range(1, 11))
y = [i**2 for i in x]


plt.plot(x, y, 'bo--', label='Квадраты чисел')  # Пунктирная линия (--), синие кружки (bo)


plt.title("График квадратов чисел")
plt.xlabel("Числа")
plt.ylabel("Квадраты чисел")


plt.show()
```

```python
import matplotlib.pyplot as plt

x = list(range(1, 11))
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x, y1, 'b^-', label='Квадраты чисел')
plt.plot(x, y2, 'ro-', label='Кубы чисел')

plt.title("Квадраты и кубы чисел")
plt.xlabel("Числа")
plt.ylabel("Значения")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Гистограмма случайных значений")
plt.xlabel("Значения")
plt.ylabel("Частота")

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Продукты', 'Транспорт', 'Развлечения', 'Образование', 'Прочее']
sizes = [30, 15, 25, 10, 20]


plt.pie(sizes, labels=None, autopct='%1.1f%%', startangle=90)
plt.legend(labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Распределение бюджета")

plt.show()

```
