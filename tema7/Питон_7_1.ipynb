```python
import matplotlib.pyplot as plt


x = list(range(1, 11))
y = [i**2 for i in x]


plt.plot(x, y, 'bo--', label='Квадраты чисел')  # Пунктирная линия (--), синие кружки (bo)


plt.title("График квадратов чисел")
plt.xlabel("Числа")
plt.ylabel("Квадраты чисел")


plt.show()
```
